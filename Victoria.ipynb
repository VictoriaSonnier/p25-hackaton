## Reachability
Implementing a function that checks whether a given board is reachable from the goal board

A permutation is said to be even if it can be obtained by an even number of successive exchanges of two squares, adjacent or not, empty or not.

We also associate a parity with the empty square: the empty square is even if we can get from the initial position of the empty square to the final position in an even number of moves, odd otherwise.

The problem will be solvable if the parity of the permutation is identical to the parity of the empty cell.

```python
def is_even(list):
    n = int(len(list)**0.5)
    tiles = [t for t in list if t != 0]
    inversions = 0
    for i in range(len(tiles)):
        for j in range(i + 1, len(tiles)):
            if tiles[i] > tiles[j]:
                inversions += 1
    
    return inversions % 2 == 0

def is_solvable(list):
    permutation_parity = is_even(list)
    ind_zero = list.index(0)
    curr_row, curr_col = divmod(ind_zero, 3)
    target_row, target_col = 2, 2
    dist_manhattan = abs(target_row - curr_row) + abs(target_col - curr_col)
    empty_cell_parity = 1-dist_manhattan % 2 #False =0, True=1
    
    return permutation_parity == empty_cell_parity


liste = [0, 2, 3, 4, 5, 1, 7, 8, 6]

# Ici, 8 et 7 sont inversés (1 inversion = impair). 
# Le 0 est à sa place (distance 0 = pair).
# Résultat attendu : False 

print(is_solvable(liste)) 
```
